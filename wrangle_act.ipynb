#### `df_twitter_archive_enhanced_clean`: Create dog_stage a categorical variable for doggo, floofer, pupper, puppo and None it does not have a value.

##### Define
create a new variable dog_stage and delete all other columns: doggo, floofer, pupper, puppo

##### Code

```python
#create an array with all dog stages
dog_stage_array = []
for x in range(df_twitter_archive_enhanced_clean['doggo'].size):
    if df_twitter_archive_enhanced_clean['doggo'].iloc[x] != 'None':
        dog_stage_array.append('doggo')
    elif df_twitter_archive_enhanced_clean['floofer'].iloc[x] != 'None':
        dog_stage_array.append('floofer')
    elif df_twitter_archive_enhanced_clean['pupper'].iloc[x] != 'None':
        dog_stage_array.append('pupper')
    elif df_twitter_archive_enhanced_clean['puppo'].iloc[x] != 'None':
        dog_stage_array.append('puppo')
    else :
        dog_stage_array.append('None')
```

```python
#create new column and add it to the dataframe
df_twitter_archive_enhanced_clean['dog_stage'] = pd.Series(dog_stage_array, index=df_twitter_archive_enhanced.index)

# To category
df_twitter_archive_enhanced_clean.dog_stage = df_twitter_archive_enhanced_clean.dog_stage.astype('category')
```

```python
#delete unnecessary columns doggo, floofer, pupper, puppo
df_twitter_archive_enhanced_clean = df_twitter_archive_enhanced_clean.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1) 
```
